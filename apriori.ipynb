```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('data.csv')
```

```
df
```

```
basket = pd.pivot_table(df , index = 'InvoiceNo' , columns ='Description' , values='Quantity' , fill_value=0)
```

```
basket.shape
```

```
basket
```

```
def units_encode(x):
  if x <=0:
    return 0
  if x >= 1:
    return 1  
```

```
basket_1 = basket.applymap(units_encode)
```

```
from mlxtend.frequent_patterns import apriori, association_rules
```

```
my_frequent = apriori(basket_1 , min_support= 0.05 , use_colnames=True)
```

```
my_frequent.head()
```

```
my_rules = association_rules(my_frequent , metric='lift' , min_threshold=1)
```

```
my_rules.head(100)
```

```
my_rules[ (my_rules['confidence'] > 0.6) & (my_rules['lift'] > 0.4 ) ]
```
